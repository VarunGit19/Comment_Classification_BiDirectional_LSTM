```python
train_df = pd.read_csv("./train.csv/train.csv")

train_df.head()
```

```python
train_df.identity_hate.value_counts()

print(train_df[train_df['identity_hate'] == 1])
```

```python
import matplotlib.pyplot as plt

counts = train_df['identity_hate'].value_counts()

counts.plot(kind='bar')
plt.title('Distribution of identity_hate')
plt.xlabel('identity_hate (0 = no, 1 = yes)')
plt.ylabel('Number of comments')
plt.xticks(rotation=0)
plt.show()
```

```python
for col in train_df.columns:
    zeros = (train_df[col] == 0).sum()
    ones = (train_df[col] == 1).sum()
    print(f"{col}: 0={zeros}, 1={ones}")
```
